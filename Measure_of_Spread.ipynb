# Measures of Spread  

Measures of spread gives us an idea of how much the data is spread or varying. Some important measures of spread are :  

1. Range  
2. Interquartile Range (IQR)  
3. Variance  
4. Standard Deviation

**1.Range**

The range describe the difference between the largest and the smallest points in your data.

```python
r=np.max(iris_Setosa['PL'])-np.min(iris_Setosa['PL'])
print("The range is",r)
```

**2.Percentiles**

A percentile is a measure used in statistics indicating the value below which a given percentage of observation in a group of observation falls.For example ,the 20th percentile is the below which 20% of the observation may be found.


```python
mylist=[1,2,3,4,5,6,7,8,9]
print(np.percentile(mylist,25))
print(np.percentile(mylist,90))
```

```python
np.percentile(iris_Setosa['PL'],95)
```

# 3. Quartiles  

Quartiles are the values that divide a list of numbers into quarters:  

Put the list of numbers in order  
Then cut the list into four equal parts  
The Quartiles are at the "cuts"  

2, 4(Q1), 4, 5(Q2), 6, 7(Q3), 8  

Q1  
lower quartile  

Q2  
middle quartile (median)  

Q3  
upper quartile

```python
print(np.percentile(iris_Setosa['PL'],25))
print(np.percentile(iris_Setosa['PL'],50))
print(np.percentile(iris_Setosa['PL'],75))
print(np.percentile(iris_Setosa['PL'],100))
```

```python
print(np.quantile(iris_Setosa['PL'],0.25))
print(np.quantile(iris_Setosa['PL'],0.50))
print(np.quantile(iris_Setosa['PL'],0.75))
print(np.quantile(iris_Setosa['PL'],1))
```

Here are the main **problems with variance** explained simply:

1. **Not intuitive**
   Variance measures squared deviations from the mean. Because of the squaring, the unit is not the same as the original data (e.g., if data is in cm, variance is in cm), which makes it less interpretable.

2. **Sensitive to outliers**
   Just like the mean, variance can be heavily influenced by extreme values. A single very large or very small number increases variance disproportionately.

3. **Not directly comparable**
   Since its in squared units, comparing variance across different datasets isnt always meaningful unless the units are the same.

4. **Harder to explain practically**
   People usually care about average spread in the same units as the data. Thats why **standard deviation** is preferred, since its the square root of variance and has the same unit as the data.

 In short: Variance is mathematically useful but less intuitive and more affected by outliers compared to other measures of spread.
